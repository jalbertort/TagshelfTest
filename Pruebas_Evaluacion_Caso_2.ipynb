```python
# Convertidor de datos de int a hex

def int2hex(num):
    '''INT2HEX :- Esta funci√≥n convierte los datos int en hex'''
    calc = hex(num)[2:].zfill(2)
    result = "0x" + calc
    return result
```

```python
# Creando funcion para secuencia de Moser-de Bruijn
def moserCode(n):
    return sum(((n >> 2 * i) & 1) << i for i in range((n.bit_length() + 1) // 2))

result = [moserCode(n) for n in range(100)]
print(result)
```
